```python
import numpy as np
import matplotlib.pyplot as plt

dt = 0.01 # sampling interval
tf = 10   # final time
n = int(tf/dt)

g = 9.8
m = 1

v0 = 0
h0 = 990

x0 = np.array([h0,v0])
u0 = 9.8*1
X = np.zeros((2,n))
U = u0*np.ones(n-1)
D = np.zeros(n-1)
X[:,0] = x0

t = np.arange(0, tf, dt)

Xdot_p = 0
hdot = 0

for k in range(n-1):
    h,v = X[:,k]
    f = 9.8 - 0.2*(h-1000) - 0.8*hdot
    d = D[k]
    U[k] = f
    hdot = v
    vdot = -g + f/m + d/m
    Xdot = np.array([hdot,vdot])
    X[:,k+1] = X[:,k] + 0.5*(3*Xdot-Xdot_p)*dt
    Xdot_p = Xdot

# h, v
plt.figure(dpi=100)
plt.subplot(211)
plt.plot(t,X[0,:])
plt.grid()
plt.subplot(212)
plt.plot(t,X[1,:])
plt.grid()
plt.show()

# f(N), d(N)
plt.figure(dpi=100)
plt.subplot(211)
plt.plot(t[:-1],U)
plt.grid()
plt.subplot(212)
plt.plot(t[:-1],D)
plt.grid()
plt.show()
```
